```python
# NUMPY cercate di trovare la media degli importi massimi pagati ogni mese a tutti i dipendenti.
import numpy as np
import pandas as pd

salario_gabri = [2000, 1900, 1950]
salario_ange = [2100, 2200, 2300]
salario_lore = [2500, 2600, 2700]

salario_totale = np.array([salario_gabri, salario_ange, salario_lore])
print(salario_totale,"\n")

bonus_gabri = [100, 200, 300]
bonus_ange = [150, 250, 350]
bonus_lore = [200, 300, 400]
bonus_totale = np.array([bonus_gabri, bonus_ange, bonus_lore])
print(bonus_totale,"\n")

salario_bonus = salario_totale + bonus_totale
print(type(salario_bonus),"\n")
print(salario_bonus,"\n")

# Trova gli importi massimi per ogni mese
massimi_mensili = np.max(salario_bonus, axis=0)
print("Importi massimi per ciascun mese:", massimi_mensili)

# Calcola la media di questi massimi
media_massimi = np.mean(massimi_mensili)
print("La media degli importi massimi pagati ogni mese a tutti i dipendenti è:", media_massimi)

#print("La media degli importi massimi pagati ogni mese a tutti i dipendenti è: ", salario_bonus.mean(axis=0,))
#print("La media degli importi massimi pagati ogni mese a tutti i dipendenti è: ", np.average(salario_bonus, axis=1))
```

```python
# PANDAS 
import pandas as pd

data = ['lore', 'ange', 'gabri']
emps_names = pd.Series(data)
print(emps_names,"\n")

data = ['lore', 'ange', 'gabri']
emps_names = pd.Series(data,index=[91,92,93])
print(emps_names,"\n")

print(emps_names[91],"\n")
print(emps_names.loc[91],"\n")
print(emps_names.iloc[0],"\n")
print(emps_names.loc[91:93],"\n")

print(emps_names.iloc[0:2],"\n")
print(emps_names[0:2],"\n")

# DataFrame
data = ['lore.pour', 'ange.bonhour', 'gabri.greco']
emps_emails = pd.Series(data,index=[91,92,93], name='emails')
emps_names.name = 'names'
df = pd.concat([emps_names, emps_emails], axis=1)
print(df,"\n")

#ESERCIZIO provate a creare un DataFrame da tre Series.

num = [1,2,3]
num = pd.Series(num, index=[91,92,93], name='num')
data = ['lore.pour', 'ange.bonhour', 'gabri.greco']
emps_emails = pd.Series(data,index=[91,92,93], name='emails')
emps_names = pd.Series(data,index=[91,92,93], name='names')
df = pd.concat([emps_names, emps_emails, num], axis=1)
print(df,"\n")
```

```python
# Dataframe pandas da API di Yfinance
import yfinance as yf
import pandas as pd
import json

tkr = yf.Ticker('TSLA')
hist = tkr.history(period="5d")
hist = hist.drop("Dividends", axis = 1)
hist = hist.drop("Stock Splits", axis = 1)
# per impostare index numerico
hist = hist.reset_index()

# per impostare la colonna data come indice
hist = hist.set_index('Date')

data = [
{"Empno":9001,"Salary":3000},
{"Empno":9002,"Salary":2800},
{"Empno":9003,"Salary":2500}
]
json_data = json.dumps(data)
salary = pd.read_json(json_data)
salary = salary.set_index('Empno')
print(salary)

#come creare un DataFrame da una lista di liste
data = [['9001','Jeff Russell', 'sales'],
['9002','Jane Boorman', 'sales'],
['9003','Tom Heints', 'sales']]
emps = pd.DataFrame(data, columns = ['Empno', 'Name', 'Job'])
column_types = {'Empno': int, 'Name': str, 'Job': str}
emps = emps.astype(column_types)
emps = emps.set_index('Empno')
print(emps)

# unire i due DataFrame
emps_salary = emps.join(salary)
print(emps_salary,f'\n')

# aggiungere una riga al DataFrame anche se manca il valore di un campo, uscirà NaN
new_emp = pd.Series({'Name': 'John Hardy', 'Job': 'sales'}, name = 9004)
emps = pd.concat([emps, new_emp.to_frame().T]) # nuovo metedo al posto di append in pandas 2...
print(emps,f'\n')
emps_salary = emps.join(salary)
print(emps_salary,f'\n')
```
